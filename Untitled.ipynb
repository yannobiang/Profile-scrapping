```python
image = cv2.imread("secret.jpeg")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite("temp/index_gray.png", gray)
```

```python
blur = cv2.GaussianBlur(gray, (7,7), 0)
```

```python
cv2.imwrite("temp/index_blur.png", blur)
```

```python
thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV)[1]
```

```python
#_INV = cv2.THRESH_OTSU
cv2.imwrite("temp/index_thresh.png", thresh)
```

```python
kernal = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 13))
```

```python
cv2.imwrite("temp/index_kernal.png", kernal)
```

```python
dilate = cv2.dilate(thresh, kernal, iterations = 1)
```

```python
cv2.imwrite("temp/index_dilate.png", dilate)
```

```python
cnts = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
```

```python
cnts = sorted(cnts, key=lambda x : cv2.boundingRect(x)[0])
```

```python
for c in cnts:
  
    x, y, w, h = cv2.boundingRect(c)
    #print(x, y, w, h)
    if h > 0 and w > 0 :
        #roi = image[y:y+h, x:x+h]
        #print(roi)
        roi = cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 10), 2)
        
        ocr_result = PT.image_to_string(roi)
        ocr_result = ocr_result.split("\n")
        for itemp in ocr_result:
            item = itemp.strip()
            print(item)
    
```
